```python
import numpy as np

def f(x):
    return np.exp(-x**2)

a, b = 0, 1
n = 10

x = np.linspace(a, b, n + 1)
h = (b - a) / n
y = f(x)

trap_integral = h * (0.5 * y[0] + np.sum(y[1:-1]) + 0.5 * y[-1])

print(f"Trapezoidal Rule Approximation: {trap_integral:.6f}")
```

```python
import numpy as np

h = 0.1
num_steps = 10

x0 = 0.0
y0 = 1.0

xs = [x0]
ys = [y0]
def f(x):
    return np.exp(-x**2)
for k in range(num_steps):
    x_current = xs[-1]
    y_current = ys[-1]
    dydx = f(x_current)
    y_next = y_current + h * dydx
    x_next = x_current + h
    xs.append(x_next)
    ys.append(y_next)

print(xs)
print(ys)
```
